# **Data Cleaning with Pandas: A Journey to Clean Data**

*Welcome to this notebook! Here, we'll guide you through the steps of cleaning your data so it’s ready for analysis. Let’s get started with loading and exploring our dataset.*

```python
import pandas as pd
import numpy as np

df = pd.read_csv('netflix_titles_2021.csv')
df.head()  
```

## **Exploring the Data**

*Let's dive into our dataset to see what we're working with. We'll check the number of rows and columns, and gather some summary statistics.*


```python
print(f"Netflix_titles Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Check for column data types and null values
df.info()

# Check summary statistics
df.describe()
```

## **Handling Missing Values**

```python
print("Missing values per column:")
print(df.isnull().sum())
```

*It looks like some data might be missing. We’ll remove rows with missing data.*

```python
# Strip leading and trailing spaces from column names
df.columns = df.columns.str.strip()

# Drop rows/columns with missing values
df.dropna(subset=['director'], inplace=True)  
df.dropna(subset=['cast'], inplace=True)
df.dropna(subset=['country'], inplace=True)
df.dropna(subset=['date_added'], inplace=True)
df.dropna(subset=['rating'], inplace=True)
df.dropna(subset=['duration'], inplace=True)

```

## Handling Duplicates

Let’s check for duplicate rows. It’s common to have duplicate entries in the data, but we can remove them to avoid skewing our analysis.


```python
# Check for duplicates
print(f"Number of duplicate rows: {df.duplicated().sum()}")

# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Check for column data types and null values
df.info()
```

## Saving the Cleaned Data

All cleaned up! Now, let’s save our dataset to a new file so we can continue our analysis without the mess. 🎉


```python
df.to_csv('cleaned_netflix_titles.csv', index=False)
```
